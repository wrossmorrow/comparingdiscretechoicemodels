```python
import pandas as pd
import numpy as np
```

```python
data = np.zeros( (10,3*5+3) )
data[:,:3*5] = np.random.randint( 1 , high=6 , size=(10,3*5) )
data[:,3*5:] = np.random.randint( 1 , high=11 , size=(10,3) )

cols = [ " " for i in range(3*5+3) ]
for i in range(3) : 
    for j in range(5) : 
        cols[ 5*i + j ] = "r_%i_%i" % (i+1,j+1) 
    cols[ 3*5 + i ] = "id_%i" % (i+1)

df = pd.DataFrame( data )
df.columns = cols
df.reset_index( inplace=True )
df.rename( columns={'index':'ind'} , inplace=True )

df
```

```python

for i in range(3) : 
    vals = [ 'r_%i_%i' % (i+1,j+1) for j in range(5) ]
    pt = pd.pivot_table( df , values=vals , index=[ 'id_%i' % (i+1) ] , columns=[] , aggfunc=np.mean)
    print( pt )
```
